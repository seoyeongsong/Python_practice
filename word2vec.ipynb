## Culture

```python
# 구 Twitter 분석기를 사용
# 결과가 잘 나오는 분석기를 사용하면 될듯함
from konlpy.tag import Okt  
okt=Okt()
# 크롤링한 파일을 열어준다.
fread = open('culture_datgul.csv', encoding="utf-8")

# 파일을 한 줄씩 읽어 토큰화를 시켜준 뒤, 특정 품사인 것만 모아서
# 훈련시킬 데이터를 만든다.

n=0
result = []

while True:
    line = fread.readline()
    
    if not line: 
        break 
    n=n+1
    
    if n%500==0:
        print("%d번째 While문."%n)
    
    tokenlist = okt.pos(line, stem=True, norm=True)
    temp=[]
    
    for word in tokenlist:
        if word[1] in ["Noun", "Adjective", "Adverb"]:
            temp.append((word[0]))
            
    # 명사나 형용사가 없는 데이터는 생략하고 있는 것만 result로 담기
    if temp: 
        result.append(temp)
fread.close()
```

```python
len(result)
```

```python
result[1]
```

```python
import warnings
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')

from gensim.models import Word2Vec
```

```python
model1 = Word2Vec(result, 
                  size=100, 
                  window=5, 
                  min_count=50, 
                  workers=4,
                  iter=100, 
                  sg=1)

# 포스태깅된 컨텐츠를 100차원의 벡터로 바꿔라. 
# 주변 단어(window)는 앞뒤로 5개 까지 보되, 
# 코퍼스 내 출현 빈도가 50번 미만인 단어는 분석에서 제외해라. 
# CPU는 쿼드코어를 쓰고 100번 반복 학습해라. 
# 분석방법론은 CBOW와 Skip-Gram 중 후자를 선택해라.

# 상당히 시간이 걸립니다...!
```

```python
test1 = []
a=model1.wv.most_similar("노인", topn=100)
for i in range(100):
    if len(a[i][0]) != 1:
        test1.append(a[i])
    else: pass
print(test1)
print(len(test1))
```
