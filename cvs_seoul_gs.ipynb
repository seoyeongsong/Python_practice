## 서울시 편의점 데이터

```python
cvs = seoul[seoul['상권업종소분류명']=="편의점"]
```

```python
cvs = cvs.reset_index()
```

```python
cvs.shape
```

```python
cvs = cvs.drop(['index'], axis=1)
```

```python
# 상호명에 25가 없을 경우 -1이 나온다.
cvs['상호명'][2].find("25")
```

## GS25 편의점만 추출하기

```python
gs_name = []
address = []
lat = []
lng = []

for i in range(len(cvs.index)):
    # 상호명에 25가 없는 데이터 중에
    if cvs['상호명'][i].find("25") == -1:
        
        
        if "지에스" not in cvs['상호명'][i]:
            pass

            if "GS" not in cvs['상호명'][i]:
                pass

                if "gs" not in cvs['상호명'][i]:
                    pass

                else:
                    gs_name.append(cvs['상호명'][i])
                    address.append(cvs['지번주소'][i])
                    lat.append(cvs['위도'][i])
                    lng.append(cvs['경도'][i])
                    print((i, cvs['상호명'][i]))
                    

            else:
                gs_name.append(cvs['상호명'][i])
                address.append(cvs['지번주소'][i])
                lat.append(cvs['위도'][i])
                lng.append(cvs['경도'][i])
                print((i, cvs['상호명'][i]))
            
    # 상호명에 25가 있는 경우
    else:

        if "지에스" in cvs['상호명'][i]:
            gs_name.append(cvs['상호명'][i])
            address.append(cvs['지번주소'][i])
            lat.append(cvs['위도'][i])
            lng.append(cvs['경도'][i])
            print((i, cvs['상호명'][i]))
            
        elif "GS" in cvs['상호명'][i]:
            gs_name.append(cvs['상호명'][i])
            address.append(cvs['지번주소'][i])
            lat.append(cvs['위도'][i])
            lng.append(cvs['경도'][i])
            print((i, cvs['상호명'][i]))
            
        elif "gs" in cvs['상호명'][i]:
            gs_name.append(cvs['상호명'][i])
            address.append(cvs['지번주소'][i])
            lat.append(cvs['위도'][i])
            lng.append(cvs['경도'][i])
            print((i, cvs['상호명'][i]))
```

```python
# 서울시 편의점 9357개 중 2935개가 gs25이다.
len(gs_name)
```

```python
dictionary =  {"상호명" : gs_name, "주소" : address, "위도" : lat, "경도" : lng}
```

```python
gs_data = pd.DataFrame(dictionary)
```

## 구 이름만 추출하여 column 추가

```python
# 주소로부터 구 이름 추출하기
gs_data['주소'][0].split()[1]
```

```python
Gu = []
for i in range(len(gs_data)):
    gu = gs_data['주소'][i].split()[1]
    Gu.append(gu)
```

```python
len(Gu)
```

```python
gs_data['구'] = Gu
```

## 지도에 시각화

```python
import folium
import json

geo_path = 'skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
m = folium.Map(location=[gs_data['위도'].mean(), gs_data['경도'].mean()], zoom_start=11, tiles="cartodbdark_matter")
```

```python
m.choropleth(geo_data=geo_str,
              data=gu_gs,
              columns=[gu_gs.index, gu_gs],
              fill_color='PuRd',
              fill_opacity=0.7,
              line_opacity=0.2,
              key_on='feature.properties.name'
              )
```
